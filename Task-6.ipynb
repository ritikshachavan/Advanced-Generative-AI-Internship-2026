```python
#Smart Parking Lot Management System


def smart_parking(capacity, vehicle_logs):
    parked = 0
    peak_usage = 0

    # Loop through vehicle logs
    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            if parked > 0:
                parked -= 1

        # Track peak usage
        if parked > peak_usage:
            peak_usage = parked

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak_usage)
    # Check parking status
    if parked > capacity:
        print("Alert: Parking Exceeded Capacity")
    else:
        print("Parking Status: Available")
#Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

```python
#Online Food Delivery Time Estimator


def delivery_time(distance, traffic, weather):

    # Assuming 5 minutes per km
    baseTime = distance * 5
    delay = 0

    # Traffic delay conditions
    if traffic.lower()=="low":
        delay += 5
    elif traffic.lower()=="medium":
        delay += 10
    elif traffic.lower()=="high":
        delay += 15

    # Weather delay conditions
    if weather.lower()=="clear":
        delay += 0
    elif weather.lower()=="rainy":
        delay += 10
    elif weather.lower()=="stormy":
        delay += 20

    # ETA calculation
    finalTime = baseTime + delay
    print("Estimated Delivery Time:", finalTime, "minutes")

#Input
distance = 8
traffic = "High"
weather = "Rainy"

delivery_time(distance, traffic, weather)
```

```python
#Movie Theatre Seat Occupancy Analyzer


def seat_occupancy(total_seats, booked_seats):

    #total booked seats using list length
    bookedCount = len(booked_seats)

    # Calculating occupancy percentage
    occupancyPercentage = (bookedCount / total_seats) * 100

    print("Occupancy:", int(occupancyPercentage), "%")

    # show status
    if occupancyPercentage == 100:
        print("Show Status: Housefull")
    elif occupancyPercentage >= 70:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional shows")
    else:
        print("Show Status: Seats Available")


#Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

seat_occupancy(total_seats, booked_seats)
```

```python
# Cloud Server Load Classification System


def server_load(CPU_readings):
    total = 0

    # calculate total CPU usage
    for reading in CPU_readings:
        total += reading

    # Average calculation
    averageLoad = total / len(CPU_readings)

    print("Average CPU Load:", int(averageLoad), "%")

    # Classification conditions
    if averageLoad < 50:
        print("Server Status: Normal")
    elif 50 <= averageLoad <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


#Input
CPU_readings = [45, 60, 70, 85, 90]

server_load(CPU_readings)
```

```python
#Smart Classroom Resource Usage Monitor

def resource_usage_monitor(resource_usage):
    overused_resources = []

    # Loop through dictionary
    for resource, hours in resource_usage.items():

        # Condition for overuse
        if hours > 8:
            overused_resources.append(resource)

    # Displaying overused resources
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


#Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_usage_monitor(resource_usage)
```

```python
# Online Event Registration Capacity Controller

def event_registration(capacity, registrations):

    confirmed = 0
    waitlisted = 0

    # processing registrations
    for i in range(registrations):

        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Display results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


#Input
capacity = 100
registrations = 105

event_registration(capacity, registrations)
```
