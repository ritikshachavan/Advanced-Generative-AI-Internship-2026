```python
#Problem 1: Smart Light Controller

class SmartLight:

    # Constructor
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    # Method to turn ON the light
    def ON(self):
        self.status = "ON"
        print(f"{self.light_name} is {self.status}")

    # Method to turn OFF the light
    def OFF(self):
        self.status = "OFF"
        print(f"{self.light_name} is {self.status}")


light_name = input("Enter Light Name: ")
action = input("Enter Action: ").upper()

# Creating object
light = SmartLight(light_name)

if action == "ON":
    light.ON()
elif action == "OFF":
    light.OFF()
else:
    print("Invalid Action!")
```

```python
#Problem 2: Employee ID Card System

class Employee:

    # Constructor
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card
    def Employee_id_card(self):
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
department = input("Enter Department: ")

# Creating object
emp = Employee(name, emp_id, department)

# Displaying ID Card
emp.Employee_id_card()
```

```python
#Problem 3: Mobile Contact Record

class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact
    def display_contact(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

# Creating object
con = Contact(name, phone)

# Displaying contact information
con.display_contact()
```

```python
#Problem 4: Product Price Tag Generator


class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    # Method to print formatted price tag
    def price_tag(self):
        print("\nProduct:", self.name)
        print(f"Price: ₹{self.price}")

name = input("Enter Product Name: ")
price = input("Enter Price: ")

# Creating object
p = Product(name, price)

# Printing price tag
p.price_tag()
```

```python
# Problem 5: Movie Rating Display System

class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = float(rating)

    # Method to display movie details
    def display_movie(self):
        print("\n Movie:", self.name)
        print(f"Rating: {self.rating} / 5")

name = input("Enter Movie Name: ")
rating = input("Enter Rating: ")

# Creating object
m = Movie(name, rating)

# Displaying movie details
m.display_movie()
```

```python
# Problem 6: Delivery Address Manager


class Delivery:

    def __init__(self, name, address):
        self.customer_name = name
        self.address = address

    # Method to display delivery details
    def display_details(self):
        print("\nDelivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

name = input("Enter Customer Name: ")
address = input("Enter Address: ")

# Creating object
d = Delivery(name, address)

# Displaying delivery information
d.display_details()
```
