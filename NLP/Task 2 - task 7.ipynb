```python
#Task 2

import re

def preprocess_text(text):

    # 1. Convert text to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove special characters (keep only alphabets and space)
    text = re.sub(r'[^a-z\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Remove very short tokens (length ≤ 2, except 'no', 'not')
    words = text.split()
    filtered_words = [word for word in words if len(word) > 2 or word in ['no', 'not']]

    text = ' '.join(filtered_words)

    return text
```

```python
# Task 5

import re
from collections import Counter

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    filtered_words = [word for word in words if len(word) > 2 or word in ['no', 'not']]

    return filtered_words


# Test Sentences
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]


#Combine all tokens
all_tokens = []

for text in test_sentences:
    tokens = preprocess_text(text)
    all_tokens.extend(tokens)


# Frequency Analysis
word_freq = Counter(all_tokens)

# Top 10 most frequent words
top_10 = word_freq.most_common(10)

# Top 5 least frequent words
least_5 = word_freq.most_common()[:-6:-1]


#Output
print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(word, ":", count)

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(word, ":", count)



```

```python
# Task 6

import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    filtered_words = [word for word in words if len(word) > 2 or word in ['no', 'not']]

    return filtered_words, ' '.join(filtered_words)


# Full Pipeline Function
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)

        all_tokens.extend(tokens)        # combine all tokens
        clean_sentences.append(clean_text)  # store cleaned sentence

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Test Input
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Visit https://openai.com now!"
]

# Run Pipeline
result = full_pipeline(test_sentences)

print("All Tokens:", result["tokens"])
print("\nClean Sentences:")
for sentence in result["clean_sentences"]:
    print(sentence)
```

```python
# Task 7

import re

def preprocess_text(text):

    #Handle empty input
    if not text or text.strip() == "":
        return [], ""

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    #Handle case where nothing remains after cleaning
    if text == "":
        return [], ""

    words = text.split()
    filtered_words = [word for word in words if len(word) > 2 or word in ['no', 'not']]

    return filtered_words, ' '.join(filtered_words)


# Full Pipeline with Error Handling
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)

        # Only add if valid
        if tokens:
            all_tokens.extend(tokens)
            clean_sentences.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


#Test Cases
test_cases = [
    "",                     # Empty string
    "😍😍😂😂🔥🔥",          # Only emojis
    "1234567890",           # Only numbers
    "Valid input text!!!"   # Normal case
]


result = full_pipeline(test_cases)

print("Tokens:", result["tokens"])
print("Clean Sentences:", result["clean_sentences"])
```
