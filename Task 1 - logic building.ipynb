```python
# 1. User Login Check


#Predefined Credentials
username = "admin"
password = "1234"


#Taking Input from User
user= input("Enter Username: ")
pwd = input("Enter Password: ")


#Checking Credentials
if(user==username and pwd ==password):

    print("Login Successful")

else:

    print("Invalid Credintials")
```

```python
#2. Pass / Fail Analyzer


#List Of Student marks
marks = [45, 78, 90, 33, 60]

Pass_Students = 0
Fail_Students = 0

#Analyzing Result
for i in marks:
  if (i >= 50):
    Pass_Students += 1

  else:
    Fail_Students += 1

#Final result
print("Total Pass Students are: ", Pass_Students)
print("Total Fail Students are: ", Fail_Students)





```

```python
#3. Simple Data Cleaner

# Original list

names = [" Alice ", "bob", " CHARLIE "]

# Empty list to store cleaned names
cleaned_list = []


for n in names:
  n =  n.strip()         # Remove extra spaces
  n = n.lower()          #Convert to lowercase
  cleaned_list.append(n)

print("Cleaned Names List: " ,cleaned_list)

```

```python
#4. Message Length Analyzer

#Original List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
  length = len(msg)     # Find length of message
  print("Message: " ,msg)
  print("Length: ", length)

  #Flag Long Messages
  if ( length > 10):
    print("Long Message ")

  else:
    print("Short Message ")
```

```python
#5. Error Message Detector

# Original list of Logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0

for log in logs:
  if(log == "ERROR"):
    error_count +=1

#Total error Count
print("Total error count is: ", error_count)




```
